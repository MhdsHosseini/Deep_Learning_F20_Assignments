```
transformations = transforms.Compose([transforms.ToPILImage(),
                                      transforms.Resize((256,256)),
                                      transforms.ToTensor()])

imag1 = transformations(img1.permute(2,0,1))
imag2 = transformations(img2.permute(2,0,1))
imag3 = transformations(img3.permute(2,0,1))
```
